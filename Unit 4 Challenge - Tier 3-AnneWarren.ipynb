Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Yes, the DF is balanced and there are only boroughs listed under the column Boroughs
# I have already removed the empty columns containing only Nans in section 2.2

# It looks like I need a Month and Year column...
properties_molten['Year']=properties_molten['Date'].dt.year
properties_molten['Month']=properties_molten['Date'].dt.month
import calendar
months_dico = dict(enumerate(calendar.month_abbr))
properties_molten['Month'] = properties_molten['Month'].map(months_dico)

# Let's make the name shorter and therefore more practical
props=properties_molten.copy()
# Here is how it looks like
print(props.head())
print(props.info())
```

```python
# I am going to fetch a quick answer with a box plot showing all the boroughs

# My first box plot shows the actual values of  housing average price per year

print('I am going to fetch a quick answer with a box plot showing all the boroughs')
print('My first box plot shows the actual values of  housing average price per year')
all_props=props.copy()
grouped_props=all_props.groupby(['Year','Boroughs']).mean()
piv_group=pd.pivot_table(grouped_props,index='Year',columns='Boroughs',values='Average Price')
piv_group.plot(kind='box')
plt.xticks(rotation=90)
plt.title('London Boroughs: Annual Average Price Variation from 1995 to 2020') 
plt.show()

# My second box plot shows a normalized version of the box plot (min/max normalization) 
# this plots shows well what I am looking for
print('My second box plot shows a normalized version of the box plot (min/max normalization)')
print('this plots shows well what I am looking for')
norm_piv_group=(piv_group-piv_group.min())/(piv_group.max()-piv_group.min())
norm_piv_group.plot(kind='box')
plt.title('London Boroughs: Normalized Annual Average Housing Price Variation from 1995 to 2020')                   
plt.xticks(rotation=90)
plt.show()

# The greatest average increase is roughly the size of the 3rd quartile
print('The greatest average increase is roughly the size of the 3rd quartile')
norm_piv_group2=norm_piv_group.copy()
norm_piv_group2.loc['median']=norm_piv_group2.median()
norm_piv_group2.loc['Q75']=norm_piv_group2.quantile(0.75,axis=0)
norm_piv_group2.loc['diff']=norm_piv_group2.loc['Q75']-norm_piv_group2.loc['median']
norm_piv_group2.loc['diff'].plot(kind='bar')
plt.title('Focusing on the 3rd quartile')
plt.show()
norm_3=norm_piv_group2.loc['diff']
norm_3.sort_values(axis=0,inplace=True,ascending=False)
norm_3.plot(kind='bar',color='red')
plt.title('Focusing on the 3rd quartile - sorted')
plt.show()

# Extracting the top 5 boroughs
top5=norm_3[0:5].index

# How the top 5 boroughs look like on a time vs. average housing price line plot  
print('How the top 5 boroughs look like on a time vs. average housing price line plot')
piv_group.plot(y=top5)
plt.title('The Top Five London Boroughs Showing the Highest Annual Average Increase in Housing Price')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True, ncol=5)
plt.show()

# How all the boroughs show next to the top 5 on a time vs. average housing price line plot
print('How all the boroughs show next to the top 5 on a time vs. average housing price line plot')
print('3 of the top5 are kind of lost in the crowd')
piv_group.plot()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, ncol=10)
plt.title('London Boroughs: Average Increase in Housing Price')
plt.show()

# Normalized line plot
print('How all the boroughs show next to the top 5 on a time vs. normalized average housing price line plot')
norm_piv_group.plot()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, ncol=10)
plt.title('London Boroughs: Average Increase in Housing Price')
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
# Here I am going to first follow the instruction regarding building a function, then I will play with average pct_change
# Creating the function create_price_ratio and having iterate through a dataframe 
def create_price_ratio (londondf,borough,topyear,bottomyear):
    df=londondf[[borough]]
    price_ratio=(df.loc[bottomyear].values/df.loc[topyear].values)
    return round(float(price_ratio),2)

dico_priceratio={}
for borough in piv_group.columns:
    result=create_price_ratio(piv_group,borough,2018,1998)
    dico_priceratio.update({borough:result})
    
price_ratio=pd.DataFrame.from_dict(dico_priceratio,orient='index')
price_ratio.sort_values(by=[0],inplace=True,ascending=False)
price_ratio.plot(kind='bar')
plt.title('Results from the function create_price_ratio which calculate price 1998/price 2018')
plt.show()


# Let's try another picture with ratio between 2000 and 2020
dico_priceratio2={}
for borough in piv_group.columns:
    result=create_price_ratio(piv_group,borough,2020,2000)
    dico_priceratio2.update({borough:result})
    
price_ratio=pd.DataFrame.from_dict(dico_priceratio2,orient='index')
price_ratio.sort_values(by=[0],inplace=True,ascending=False)
price_ratio.plot(kind='bar',color='green')
plt.title('Results from the function create_price_ratio which calculate price 2000/price 2020')
plt.show()

# Using percent change aggregated to get a sense of growth from year to year
def expmean (londondf,borough):
    df=londondf[[borough]].pct_change()
    expme=df.sum()
    return round(float(expme),3)

normdata=norm_piv_group2.loc[[1998,1999,2000,2001,2020,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]]
regdata=piv_group
dfdata=normdata.loc[[1998,1999,2000,2001,2020,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]]

dico_expme={}
for borough in dfdata.columns:
    result2=expmean(dfdata,borough)
    dico_expme.update({borough:result2})
    
print(dico_expme)
expmep=pd.DataFrame.from_dict(dico_expme,orient='index')
expmep.sort_values(by=[0],inplace=True,ascending=False)
expmep.plot(kind='bar',color='red')
plt.title('Results showing cumulative growth from 1998 to 2018')
plt.show()

# Normalized line plot
print('How all the boroughs show next to the top 5 on a time vs. normalized average housing price line plot')
norm_piv_group.plot()
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, ncol=10)
plt.title('London Boroughs: Normalized Average Increase in Housing Price')
plt.show()
```
